```python
## Top 5 Trusted Governments Countries and their other trusted ratio
df_trusted_Gover = df.nlargest(5, 'Government')
df_trusted_Gover
```

```python
Neighbourhood = df_trusted_Gover.groupby(['Country'])['Neighbourhood'].mean()
Government = df_trusted_Gover.groupby(['Country'])['Government'].mean()
Scientist = df_trusted_Gover.groupby(['Country'])['Scientist'].mean()
Journalist = df_trusted_Gover.groupby(['Country'])['Journalist'].mean()
Doctor_and_Nurses = df_trusted_Gover.groupby(['Country'])['Doctor and Nurses'].mean()
Philantropist = df_trusted_Gover.groupby(['Country'])['Philantropist'].mean()
Traditional_Healers = df_trusted_Gover.groupby(['Country'])['Traditional Healers'].mean()
print(Neighbourhood)
print('-------------------------------------')
print(Government)
print('-------------------------------------')
print(Scientist)
print('-------------------------------------')
print(Journalist)
print('-------------------------------------')
print(Doctor_and_Nurses)
print('-------------------------------------')
print(Philantropist)
print('-------------------------------------')
print(Traditional_Healers)
```

```python
labels =df_trusted_Gover.groupby('Country').count().index
x = np.arange(len(labels))
plt.figure(figsize=(15,10))
plt.bar(x-0.25,Neighbourhood,width=0.10,label='Neighbourhood')
plt.bar(x-0.18,Government,width=0.10,label='Government')
plt.bar(x-0.10,Scientist,width=0.10,label='Scientist')
plt.bar(x,Journalist,width=0.10,label='Journalist')
plt.bar(x+0.10,Doctor_and_Nurses,width=0.10,label='Doctor_and_Nurses')
plt.bar(x+0.18,Philantropist,width=0.10,label='Philantropist')
plt.bar(x+0.25,Traditional_Healers,width=0.10,label='Traditional_Healers')
plt.legend(loc=7)
plt.xticks(x,labels,fontsize=17)
plt.title('Top 5 Trusted Governments Countries and their other trusted ratio')
```

## United State

```python
df_United_State = df[df['Country']=='United States']
df_United_State
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df_United_State)
plt.title('United State trust ratio')
```

```python
df_United_State1 = df_United_State.groupby('Country').mean()
df_United_array = np.array(df_United_State1)
df_United_array
```

```python
import random
fig = plt.figure(figsize=(8,5),dpi=100)
axes = fig.add_axes([0.1,0.1,0.9,0.9])
# pie chart
types = ['Neighbourhood' , 'Government' , 'Scientist' , 'Journalist' , 'Doctor and Nurses' , 'Philantropist' ,
         'Traditional Healers']
colors = []
for i in range(7):
    rgp = (random.uniform(0,.10),random.uniform(0,.5),random.uniform(0,.5))
    colors.append(rgp)
explode = [0] * 7
explode[4] = 0.05
wedges , texts , autotexts = plt.pie(df_United_array[0] , explode=explode ,labels = types, colors=colors , autopct='%1.0f%%'
                                    , shadow = True , startangle=140 , textprops=dict(color='w'))
plt.legend(wedges , types , loc='right' , bbox_to_anchor=(1 , 0 , .5 , 1))
plt.title('United State trust ratio')
```
